```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4],[1,2,2,4])
```

```python
x = np.linspace(0, 10, 100) #Return evenly spaced numbers over a specified interval.
fig, ax = plt.subplots()  
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label="cubic")
ax.plot(x, 100*x, label="100x")
ax.plot(x, np.ones((100,1)), label="ones")
ax.legend()
```

```python
x=np.random.random((200,1))
fig, ax = plt.subplots()
ax.plot(x, label='rand')
```

```python
def fsin(x):
    return np.sin(x)

def fcos(x):
    return np.cos(x)

t1 = np.arange(0.0, 5, 0.1)
fig = plt.figure()
plt.figure()
plt.grid(True)
plt.plot(t1, fsin(t1), t1, fcos(t1))
```

```python
x = np.arange(15)**2
plt.figure()
plt.grid(True)

plt.yscale('linear')
plt.title('linear')
plt.plot(x)
```
