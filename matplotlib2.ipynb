```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 50)
y= x**2
plt.plot(x, y, color="red", label="Graph", linestyle="dashed", marker='x')
```

```python
y= 0.5*np.sin(x)
plt.plot(x, y, color="red", label="Graph", linestyle="dashed", marker='x')
```

```python
y= np.exp(-x**2)
plt.plot(x, y, color="red", label="Graph", linestyle="dashed", marker='x')
```
